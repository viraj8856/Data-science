## Regression Model

```python
df = pd.get_dummies(df, columns=['State'])
```

```python
X = df[['R&D Spend','Administration', 'Marketing Spend', 'State_California', 'State_Florida', 'State_New York']]
Y = df[['Profit']]
```

```python
model = sm.OLS(Y, X).fit()
predictions = model.predict(X)
```

```python
model.summary()
```

## Solving Multicollinearity

```python
# Finding Cook's Distance
```

```python
inf = model.get_influence()
```

```python
inf_sf = inf.summary_frame()
```

```python
inf_sf.sort_values('cooks_d', ascending=False)
```

```python
inf.plot_influence()
```

### VIF 

```python
vif = pd.DataFrame()
```
