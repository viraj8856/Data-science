```python
df["sales"]=(df["Sales"]>7.49)*1
```

```python
label_encoder = preprocessing.LabelEncoder()
df['ShelveLoc']= label_encoder.fit_transform(df['ShelveLoc']) 
df['Urban']= label_encoder.fit_transform(df['Urban']) 
df['US']= label_encoder.fit_transform(df['US']) 
```

```python
df
```

```python
x=df.iloc[:,0:10]
y=df.iloc[:,10]
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr()>0.6, annot = True,cmap="Dark2",vmin=0.5, vmax=0.7)
plt.show()
```

```python
sns.heatmap(df.corr(),cmap="Dark2",annot=True,vmin=0.5, vmax=0.7)
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=20)
```

```python
model = DecisionTreeClassifier(criterion = 'gini',max_depth=3)
model.fit(x_train,y_train)
```

```python
fn=['CompPrice','Income','Advertising','Population','Price','ShelveLoc','Age','Education','Urban','US','income']
cn=['YES','NO']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (5,5), dpi=300)
tree.plot_tree(model,
               feature_names = fn, 
               class_names=cn,
               filled = True); 
```

```python
model.score(x_train, y_train)
```

```python
# Predicting with test data 
preds = model.predict(x_test)
```

```python
# Checking the accuracy with predicted and tested
np.mean(preds==y_test)
```

```python
#Checking the model score
model.score(x_test, y_test)
```

```python
model.feature_importances_
```

```python
y_pred = preds
('Accuracy Score:',metrics.accuracy_score(y_test,y_pred))
```

```python
# Prediction with probabilities of the classes
model.predict_proba(x_test)
```

```python
# Confusion matrix
cm= confusion_matrix(y_test, y_pred) 
cm
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(7,5))
sns.heatmap(data=cm,linewidths=.5, annot=True,square = True,cmap="Blues")
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
all_sample_title = 'Accuracy Score: {0}'.format(model.score(x_test, y_test))
plt.title(all_sample_title, size = 15)
```

```python
#Classification report - Y_test data and the predicted data
print("Classification report - \n", classification_report(y_test,y_pred))
```

```python
# identify the train and test accuracies of the prediction, with the change in the max_depth 
train_accuracy = []
test_accuracy = []
for max_depth in range(1,11):
    model = DecisionTreeClassifier(criterion='entropy', max_depth= max_depth, random_state= 40)
    model.fit(x_train, y_train)
    train_accuracy.append(model.score(x_train, y_train))
    test_accuracy.append(model.score(x_test, y_test))
```

```python
# Finding the accuracies in different maximum depth in training and testing data
accuracy = pd.DataFrame({'max_depth':range(1,11),'train_acc':train_accuracy, 'test_acc':test_accuracy})
accuracy
```

```python
labels =['train_acc','test_acc']
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot()
plt.plot(accuracy['max_depth'],accuracy['train_acc'],color='red', marker = '*')
plt.plot(accuracy['max_depth'], accuracy['test_acc'],color='green' ,marker = 'o')
ax.set_xticks(np.arange(1,11))
plt.legend(labels)
plt.xlabel('max_depth of the Decision Tree')
plt.ylabel('Performance')
plt.show()
```
